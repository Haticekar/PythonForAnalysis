## üü© Set Index and Display Rows (2.5.1)

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"]) #Convert the job_posted_date column to datetime format.
df.set_index("job_posted_date", inplace=True) #Deƒüi≈üiklik doƒürudan df √ºzerinde yapƒ±lƒ±r, yeni deƒüi≈ükene atamaya gerek kalmaz(inplace=True)
df.head()
```

---

## üü© Reset Index and Drop Existing (2.5.2)

# Ne isteniyor?
* job_posted_date artƒ±k satƒ±r indeksi olmamalƒ±.
* Bunun yerine, satƒ±rlar yeniden 0, 1, 2, 3... ≈üeklinde numaralandƒ±rƒ±lmalƒ±.
* job_posted_date s√ºtunu da tabloya geri s√ºtun olarak gelmemeli (yani tamamen silinmeli).

### drop=True
* Normalde reset_index() yaptƒ±ƒüƒ±nda, eski index olan s√ºtun (job_posted_date) veriye geri eklenir.
* Ancak biz bu s√ºtunu artƒ±k istemiyoruz, o y√ºzden drop=True diyoruz ve tamamen atƒ±yoruz.

```python
df.reset_index(drop=True, inplace=True) # DataFrame‚Äôin mevcut indeksini sƒ±fƒ±rlayƒ±p yerine varsayƒ±lan sayƒ±sal indeks (0, 1, 2, ...) koyar.
df.head()
```

---

## üü© Sort Index (2.5.3)

```python
pivot = df.pivot_table(
    index="company_name",
    values="salary_year_avg",
    aggfunc = "median"
)
pivot_sorted = pivot.sort_index(ascending=False)
pivot_sorted.head()
```

```python
grouped = df.groupby("company_name")["salary_year_avg"].median()
grouped_sorted = grouped.sort_values(ascending=False)
grouped_sorted.head()
```

```python
pivot_table = df.pivot_table(values='salary_year_avg', index='company_name', aggfunc='median')
pivot_table.sort_index(inplace=True, ascending=False)
pivot_table.head()
```

## üü® Pivot Table Reset & Set Index (2.5.4)

```python
pvt_table = df.pivot_table(
    index="job_title_short",
    values="salary_year_avg",
    aggfunc="median"
)
pvt_table.reset_index(inplace=True)
pvt_table.set_index("salary_year_avg", inplace=True)
pvt_table.head()
```
