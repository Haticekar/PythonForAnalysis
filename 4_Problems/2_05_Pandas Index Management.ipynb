## 🟩 Set Index and Display Rows (2.5.1)

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"]) #Convert the job_posted_date column to datetime format.
df.set_index("job_posted_date", inplace=True) #Değişiklik doğrudan df üzerinde yapılır, yeni değişkene atamaya gerek kalmaz(inplace=True)
df.head()
```

---

## 🟩 Reset Index and Drop Existing (2.5.2)

# Ne isteniyor?
* job_posted_date artık satır indeksi olmamalı.
* Bunun yerine, satırlar yeniden 0, 1, 2, 3... şeklinde numaralandırılmalı.
* job_posted_date sütunu da tabloya geri sütun olarak gelmemeli (yani tamamen silinmeli).

### drop=True
* Normalde reset_index() yaptığında, eski index olan sütun (job_posted_date) veriye geri eklenir.
* Ancak biz bu sütunu artık istemiyoruz, o yüzden drop=True diyoruz ve tamamen atıyoruz.

```python
df.reset_index(drop=True, inplace=True) # DataFrame’in mevcut indeksini sıfırlayıp yerine varsayılan sayısal indeks (0, 1, 2, ...) koyar.
df.head()
```

---

## 🟩 Sort Index (2.5.3)

```python
pivot = df.pivot_table(
    index="company_name",
    values="salary_year_avg",
    aggfunc = "median"
)
pivot_sorted = pivot.sort_index(ascending=False)
pivot_sorted.head()
```

```python
grouped = df.groupby("company_name")["salary_year_avg"].median()
grouped_sorted = grouped.sort_values(ascending=False)
grouped_sorted.head()
```

```python
pivot_table = df.pivot_table(values='salary_year_avg', index='company_name', aggfunc='median')
pivot_table.sort_index(inplace=True, ascending=False)
pivot_table.head()
```

---

## 🟨 Pivot Table Reset & Set Index (2.5.4)

```python
pvt_table = df.pivot_table(
    index="job_title_short",
    values="salary_year_avg",
    aggfunc="median"
)
pvt_table.reset_index(inplace=True)
pvt_table.set_index("salary_year_avg", inplace=True)
pvt_table.head()
```
