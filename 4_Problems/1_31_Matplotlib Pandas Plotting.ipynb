```python
!pip install datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Job Posting Location Bar Plot (1.31.1)

```python
title = df["job_via"].value_counts().head(5)
title.plot(kind="bar", title="Top 5 Job Posting Platforms", xlabel="Platform", ylabel="Number of Job Postings", label="count")
plt.legend()
```

---

## ðŸŸ© Job Postings Trend Line Plot (1.31.2)

```python
monthly_job_counts = pd.to_datetime(df["job_posted_date"])
monthly_job_counts = monthly_job_counts.dt.to_period("M")
monthly_job_counts = monthly_job_counts.value_counts().sort_index()
monthly_job_counts.plot(kind="line", title= "Trend of Job Postings Over Time", xlabel="Month", ylabel="Number of Job Postings")
monthly_job_counts
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df["job_posted_month"] = df["job_posted_date"].dt.to_period("M")
monthly_job_counts = df["job_posted_month"].value_counts().sort_index()
monthly_job_counts.plot(kind="line")
plt.title("Trend of Job Postings Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Job Postings")
plt.show()
```

---

## ðŸŸ¨ Top Job Locations (1.31.3)

```python
location = df["job_location"].value_counts().head(5)
location.plot(kind="barh", title="Number of Job Postings by Job Location", xlabel="Count of Jobs", ylabel="Job Location", label="count")
plt.legend()
```
