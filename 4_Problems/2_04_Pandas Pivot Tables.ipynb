```python
!pip install datasets
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

 Neden values kullanmıyoruz?

📌 pivot_table() fonksiyonu:

  df.pivot_table(index=..., columns=..., values=..., aggfunc=...)

* values → sayısal veri ister (örneğin maaş sütunu)
* aggfunc='mean', 'sum', 'max' gibi işlevler sayısal değerleri özetler

# 🧠 Açıklama:
1. Bu örnekte sayısal bir sütunu özetlemiyoruz.
2.  Sadece her kombinasyonun kaç kez geçtiğini (satır sayısını) saymak istiyoruz.
3. Bu yüzden 'values' parametresine gerek yok.
4. 'aggfunc="size"' kullanarak doğrudan satır sayısı elde edilir.

## 🟩 Count Job Postings by Title and Country (2.4.1)

```python
df.pivot_table(index=["job_title_short", "job_country"], aggfunc="size").head(10)
#sayısal değer sütunu olmadığı için value yerine sadece index kullandık
```

---

## 🟨 Salary Stats by Job Title (2.4.2)

```python
pivot = df.pivot_table(
    index="company_name",
    values="salary_year_avg",
    aggfunc=["max","mean","min"]).head()

pivot
```

---

## 🟥 Sum Salary by Title and Country (2.4.3)

```python
df_data_science = df[df["job_title_short"]=="Data Scientist"] #only data science jobs

pvt_table = df_data_science.pivot_table(
    index= ["company_name", "job_country"],
    values="salary_year_avg",
    aggfunc="median"
)
filtered = pvt_table[pvt_table["salary_year_avg"]>200000]
filtered.head(10)
```

```python
# Filter for only Data Scientist jobs
df_DS = df[df['job_title_short'] == 'Data Scientist'].copy()

# Create pivot table
pivot_table = df_DS.pivot_table(values='salary_year_avg', index=['company_name', 'job_country'], aggfunc='median')
filtered_pivot_table = pivot_table[pivot_table['salary_year_avg'] > 200000]
filtered_pivot_table.head(10)
```
