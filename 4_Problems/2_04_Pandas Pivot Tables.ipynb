```python
!pip install datasets
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

 Neden values kullanmƒ±yoruz?

üìå pivot_table() fonksiyonu:

  df.pivot_table(index=..., columns=..., values=..., aggfunc=...)

* values ‚Üí sayƒ±sal veri ister (√∂rneƒüin maa≈ü s√ºtunu)
* aggfunc='mean', 'sum', 'max' gibi i≈ülevler sayƒ±sal deƒüerleri √∂zetler

# üß† A√ßƒ±klama:
1. Bu √∂rnekte sayƒ±sal bir s√ºtunu √∂zetlemiyoruz.
2.  Sadece her kombinasyonun ka√ß kez ge√ßtiƒüini (satƒ±r sayƒ±sƒ±nƒ±) saymak istiyoruz.
3. Bu y√ºzden 'values' parametresine gerek yok.
4. 'aggfunc="size"' kullanarak doƒürudan satƒ±r sayƒ±sƒ± elde edilir.

## üü© Count Job Postings by Title and Country (2.4.1)

```python
df.pivot_table(index=["job_title_short", "job_country"], aggfunc="size").head(10)
#sayƒ±sal deƒüer s√ºtunu olmadƒ±ƒüƒ± i√ßin value yerine sadece index kullandƒ±k
```

## üü® Salary Stats by Job Title (2.4.2)

```python
pivot = df.pivot_table(
    index="company_name",
    values="salary_year_avg",
    aggfunc=["max","mean","min"]).head()

pivot
```

## üü• Sum Salary by Title and Country (2.4.3)

```python
df_data_science = df[df["job_title_short"]=="Data Scientist"] #only data science jobs

pvt_table = df_data_science.pivot_table(
    index= ["company_name", "job_country"],
    values="salary_year_avg",
    aggfunc="median"
)
filtered = pvt_table[pvt_table["salary_year_avg"]>200000]
filtered.head(10)
```

```python
# Filter for only Data Scientist jobs
df_DS = df[df['job_title_short'] == 'Data Scientist'].copy()

# Create pivot table
pivot_table = df_DS.pivot_table(values='salary_year_avg', index=['company_name', 'job_country'], aggfunc='median')
filtered_pivot_table = pivot_table[pivot_table['salary_year_avg'] > 200000]
filtered_pivot_table.head(10)
```
