```python
!pip install datasets
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## üü© Drop Missing Values (2.3.1)

```python
df_original = df.copy()
df_altered = df_original.dropna(subset=["salary_year_avg"]) # # Bu s√ºtunda NaN olan satƒ±rlar silinir
df_altered.head()
```

## üü© Random Sample of DataFrame (2.3.2)

## üü® Fill Missing Salaries (2.3.3)

```python
df_original = df.copy()
median_salary = df_original["salary_year_avg"].median()

df_altered = df_original.copy()
df_altered["salary_year_avg"].fillna(median_salary, inplace=True) # NaN deƒüerleri median ile doldurur

df_altered.head(5)
```

```python
df_copy = df.copy()
median_salary = df_copy['salary_year_avg'].median()
df_copy['salary_year_avg'] = df_copy['salary_year_avg'].fillna(median_salary)
df_copy.head(5)
```
