## üü© Extract Year from Date (1.25.3)

```python
# convert the date data to a datetime object
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
#Bu satƒ±r, job_posted_date s√ºtunundaki verileri datetime nesnesine d√∂n√º≈üt√ºr√ºr.
#Datetime nesnesi, tarihle ilgili √∂zellikler (yƒ±l, ay, g√ºn, saat vb.) ve metodlar (√∂rneƒüin, dt.year, dt.month) kullanabilmemizi saƒülar.

df["job_posted_year"] = df["job_posted_date"].dt.year
#Bu satƒ±r, job_posted_date s√ºtunundaki her bir tarihin yƒ±l bilgisini √ßƒ±karƒ±r.
#Ancak, bu i≈ülemi yapabilmek i√ßin job_posted_date s√ºtununun zaten datetime nesnesi olmasƒ± gerekir.
df.head(5)
```
