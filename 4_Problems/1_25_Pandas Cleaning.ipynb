## 🟩 Extract Year from Date (1.25.3)

```python
# convert the date data to a datetime object
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
#Bu satır, job_posted_date sütunundaki verileri datetime nesnesine dönüştürür.
#Datetime nesnesi, tarihle ilgili özellikler (yıl, ay, gün, saat vb.) ve metodlar (örneğin, dt.year, dt.month) kullanabilmemizi sağlar.

df["job_posted_year"] = df["job_posted_date"].dt.year
#Bu satır, job_posted_date sütunundaki her bir tarihin yıl bilgisini çıkarır.
#Ancak, bu işlemi yapabilmek için job_posted_date sütununun zaten datetime nesnesi olması gerekir.
df.head(5)
```
